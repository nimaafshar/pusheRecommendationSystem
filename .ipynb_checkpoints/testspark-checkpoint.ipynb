```python
# To find out where the pyspark
import findspark
findspark.init()
```

```python
# Creating Spark Context
from pyspark import SparkContext
sc
```

```python
# Calculating words count
text_file = sc.textFile("notifs.txt")
```

```python
# Calculating words count
counts = text_file.flatMap(lambda line: line.split(" ")[1:])\
                    .map(lambda word:(word,1))\
                    .reduceByKey(lambda a,b:a+b)
                    
```

```python
counts.toDF().show()
```
